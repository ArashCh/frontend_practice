```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pprint
from tqdm import tqdm
import pandas as pd
import json
```

```python
PP = pprint.PrettyPrinter(indent=4)
```

```python
URL = "https://react-projects-15-cocktails.netlify.app/"
```

```python
def run_driver(url):
    driver = webdriver.Chrome(
        "G:\Workspace\Practice and Test\WEB\Frontend\Khashayar Mahdavi\\0004 - the cocktail db\python\chromedriver.exe")
    driver.get(url)
    return driver
```

```python
driver = run_driver(URL)
```

```python
# getting the cocktails
articles = driver.find_elements_by_tag_name("article")
```

```python
cocktail_links = [article.find_element_by_tag_name("a").get_attribute("href") for article in articles]
```

```python
cocktail_data = {}
for cl in tqdm(cocktail_links):
    driver.get(cl)
    # wait till the page is fully loaded
    WebDriverWait(driver=driver, timeout=10).until(EC.element_to_be_clickable((By.CLASS_NAME, "section-title")))
    title = driver.find_element_by_class_name("section-title").text
    cocktail_data[title] = {}
    info = driver.find_element_by_class_name("drink-info").find_elements_by_tag_name("p")
    cocktail_data[title]["name"] = info[0].text
    cocktail_data[title]["category"] = info[1].text
    cocktail_data[title]["info"] = info[2].text
    cocktail_data[title]["glass"] = info[3].text
    cocktail_data[title]["instructions"] = info[4].text
    cocktail_data[title]["ingredients"] = info[5].text
    cocktail_data[title]["img_src"] = driver.find_element_by_class_name("drink").find_element_by_tag_name("img").get_attribute("src")
```

```python
PP.pprint(cocktail_data)
```

```python
# json.dump(cocktail_data, "cocktail_data.json")
with open('cocktail_data.json', 'w') as fp:
    json.dump(cocktail_data, fp)
```
